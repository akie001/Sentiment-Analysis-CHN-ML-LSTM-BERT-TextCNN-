```python
# 查看数据标签的分布

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 读取数据
train_data = pd.read_csv('data/train.tsv', sep='\t')
valid_data = pd.read_csv('data/dev.tsv', sep='\t')
test_data = pd.read_csv('data/test.tsv', sep='\t') 

# 统计label项的比例和分布
train_label = train_data['label'].value_counts()
valid_label = valid_data['label'].value_counts()
test_label = test_data['label'].value_counts()

# 绘制饼图
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.pie(train_label.values, labels=train_label.index, autopct='%1.1f%%')
plt.title('train_label')
plt.subplot(222)
plt.pie(valid_label.values, labels=valid_label.index, autopct='%1.1f%%')
plt.title('valid_label')
plt.subplot(223)
plt.pie(test_label.values, labels=test_label.index, autopct='%1.1f%%')
plt.title('test_label')
plt.savefig('data_check.png')
plt.show()
```
